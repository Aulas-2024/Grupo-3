```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
data = pd.read_csv('hsb2f.csv', sep=';')
df = data.loc[:, data.columns != 'id']
df
```

```python
# Gráfico 1: Gênero x Classe Social (contagem)
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='genero', hue='clasocial')
plt.title('Distribuição de Gênero por Classe Social')
plt.xlabel('Gênero')
plt.ylabel('Contagem')
plt.legend(title='Classe Social')
plt.show()
```

```python
# Gráfico 2: Média de Leitura e Escrita por Tipo de Escola
df['ler_escrever'] = (df['ler'] + df['escrever']) / 2
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='tipescola', y='ler_escrever')
plt.title('Média de Leitura e Escrita por Tipo de Escola')
plt.xlabel('Tipo de Escola')
plt.ylabel('Pontuação Média (Ler + Escrever)')
plt.show()
```

```python
# Gráfico 3: Gênero x Programa x Leitura e Escrita
plt.figure(figsize=(12, 8))
sns.barplot(data=df, x='genero', y='ler_escrever', hue='programa')
plt.title('Média de Leitura e Escrita por Gênero e Programa')
plt.xlabel('Gênero')
plt.ylabel('Pontuação Média (Ler + Escrever)')
plt.legend(title='Programa')
plt.show()
```

```python
# Gráfico 4: Média de Leitura, Escrita e Matemática por Tipo de Escola
# Calcular a média de Leitura, Escrita e Matemática
df['ler_escrever_matematica'] = (df['ler'] + df['escrever'] + df['matematica']) / 3

# Plot: Média de Leitura, Escrita e Matemática por Tipo de Escola, com divisão por Gênero, Classe Social e Programa
g = sns.catplot(
    data=df,
    x='tipescola',
    y='ler_escrever_matematica',
    hue='genero',
    col='clasocial',
    row='programa',
    kind='bar',
    height=4,
    aspect=1.2,
    palette='viridis'
)

g.set_axis_labels("Tipo de Escola", "Pontuação Média (Leitura, Escrita e Matemática)")
g.set_titles("{col_name} Classe Social | {row_name} Programa")
plt.suptitle("Média de Leitura, Escrita e Matemática por Tipo de Escola, Gênero, Classe Social e Programa", y=1.03)
plt.show()
```

```python
# Gráfico 3: Gênero x Programa x Ciências e Estudos Sociais
df['ciencias_estsociais'] = (df['ciencias'] + df['estsociais']) / 2
plt.figure(figsize=(14, 8))
sns.barplot(data=df, x='genero', y='ciencias_estsociais', hue='programa', palette="coolwarm")
plt.title('Média de Ciências e Estudos Sociais por Gênero e Programa')
plt.xlabel('Gênero')
plt.ylabel('Pontuação Média (Ciências e Estudos Sociais)')
plt.legend(title='Programa')
plt.show()
```

```python
# Gráfico 4: Comparação Geral por Classe Social e Programa
df['pontuacao_total'] = (df['ler'] + df['escrever'] + df['matematica'] + df['ciencias'] + df['estsociais']) / 5
plt.figure(figsize=(14, 8))
sns.barplot(data=df, x='clasocial', y='pontuacao_total', hue='programa', palette="magma")
plt.title('Pontuação Média Geral por Classe Social e Programa')
plt.xlabel('Classe Social')
plt.ylabel('Pontuação Média Geral')
plt.legend(title='Programa')
plt.show()
```
